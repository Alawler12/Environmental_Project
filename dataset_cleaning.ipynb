```python
#compare countries in each dataset
co2_names = co2_df['country'].unique().tolist()
co2_names
```

```python
# remove space before country in temps_df

temps_df.country = temps_df.country.str.lstrip()
temp_names = temps_df['country'].unique().tolist()
temp_names
```

```python
#find which country names are different between dataframes

def Diff(li1, li2):
    li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2]
    return li_dif

Diff(co2_names, temp_names)
```
