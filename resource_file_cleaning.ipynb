```python
import pandas as pd
import datetime as dt
import glob
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
# import all the separate temperature CSVs and concatenate them into one data frame
path = r'c:\users\ashle\onedrive\desktop\Classwork\Environmental_Project\Resources\temps'                      
temps_csvs = glob.glob(os.path.join(path, "*.csv"))     

df_for_each_csv = [pd.read_csv(f) for f in temps_csvs]
temps_df = pd.concat(df_for_each_csv, ignore_index = True)
```

```python
temps_df
```

```python
# drop statistics column from temps
clean_temps = temps_df.drop([' Statistics'], axis=1)
```

```python
#rename temp_df columns
temps_df = clean_temps.rename(columns={'Temperature - (Celsius)':'temperature(C)', ' Year':'year', ' Country':'country', ' ISO3':'iso_code'})
temps_df
```

```python
# remove space before country names in temps_df
temps_df.country = temps_df.country.str.lstrip()
```

```python
# reorder columns
temps_df = temps_df[['country', 'iso_code', 'year', 'temperature(C)']]
```

```python
# find average temp for each year in each country

avg_temps = temps_df.groupby(['iso_code','country','year']).mean()
clean_avg_temps = avg_temps.reset_index()
clean_avg_temps
```

```python
# sort df
clean_avg_temps = clean_avg_temps.sort_values(['country', 'year'])

clean_avg_temps
```

```python
#export as single CSV
clean_avg_temps.to_csv('global_temps_1901_2016.csv', index = False)
```

```python
# import CO2 dataset
co2_df = pd.read_csv('./Resources/owid-co2-data_updated.csv')
```

```python
co2_df 
```

```python
# remove country names that don't represent individual countries

not_country = ['Africa', 'Asia', 'Asia (excl. China & India)','EU-27','EU-28', 'Europe','Europe (excl. EU-27)',
               'Europe (excl. EU-28)', 'International transport', 'Kuwaiti Oil Fires', 'North America',
               'North America (excl. USA)', 'World', 'South America']

co2_df = co2_df.loc[co2_df['country'].apply(lambda x: x not in not_country)]
```

```python
# sort alphabetically
co2_df = co2_df.sort_values(['country', 'year'])
```

```python
#export as clean CSV
co2_df.to_csv('clean_co2_data.csv', index = False)
```
