```python
import pandas as pd
```

```python
temps = pd.read_csv('global_temps_1901_2016.csv')
```

```python
co2 = pd.read_csv('clean_co2_data.csv')
```

```python
co2_names = co2.country.unique().tolist()
co2_names
```

```python
temp_names = temps['country'].unique().tolist()
temp_names
```

```python
def Diff(li1, li2):
    li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2]
    return li_dif

Diff(co2_names, temp_names)
```
