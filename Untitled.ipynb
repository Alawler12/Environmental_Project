```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# clean temp data
temps = pd.read_csv('./Resources/global_temps_to_2016.csv')
```

```python
# additional temp data
more_temps = pd.read_csv('./Resources/temps/city_temperature.csv')
```

```python
co2 = pd.read_csv('./Resources/global_co2_data_to_2019.csv')
```

```python
# clean up new temp data and find average country temp for each year
# drop dumb columns
clean_temps = more_temps.drop(['State', 'Region', 'Month', 'Day'], axis=1)
clean_temps
```

```python
# convert temp to celcius
clean_temps['Temperature(C)'] = (clean_temps['AvgTemperature'] - 32) * .5556
clean_temps
```

```python
# drop avgtemp column
more_temps = clean_temps.drop(['AvgTemperature'], axis=1)
more_temps
```

```python
# find average temp for country by year
city_avg = more_temps.groupby(['Country','City','Year']).mean()
city_avg_temp = city_avg.reset_index()
city_avg_temp
```

```python
# rename columns
more_temps.columns = map(str.lower, clean_temps.columns)
more_temps
```

```python
co2_names = co2['country'].unique().tolist()
```

```python
temp_names = temps['country'].unique().tolist()
```

```python
more_temp_names = more_temps['Country'].unique().tolist()
temp_names
```

```python
def Diff(li1, li2):
    li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2]
    return li_dif

Diff(more_temp_names, temp_names)
```
