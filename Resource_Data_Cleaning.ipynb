```python
import pandas as pd
import datetime as dt
import glob
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
# import all the separate temperature CSVs and concatenate them into one data frame
path = r'c:\users\ashle\onedrive\desktop\Classwork\Environmental_Project\Resources\temps'                      
temps_csvs = glob.glob(os.path.join(path, "*.csv"))     

df_for_each_csv = [pd.read_csv(f) for f in temps_csvs]
temps_df = pd.concat(df_for_each_csv, ignore_index = True)
```

```python
temps_df
```

```python
# drop statistics column from temps
clean_temps = temps_df.drop([' Statistics'], axis=1)
```

```python
#rename temp_df columns
temps_df = clean_temps.rename(columns={'Temperature - (Celsius)':'temperature(C)', ' Year':'year', ' Country':'country', ' ISO3':'iso_code'})
temps_df
```

```python
# remove space before country names in temps_df
temps_df.country = temps_df.country.str.lstrip()
```

```python
# reorder columns
temps_df = temps_df[['country', 'iso_code', 'year', 'temperature(C)']]
```

```python
# find average temp for each year in each country

avg_temps = temps_df.groupby(['iso_code','country','year']).mean()
clean_avg_temps = avg_temps.reset_index()
clean_avg_temps
```

```python
# sort df alphabetically
clean_avg_temps = clean_avg_temps.sort_values(['country', 'year'])

clean_avg_temps
```

```python
#export as single CSV
clean_avg_temps.to_csv('global_temps_1901_2016.csv', index = False)
```

```python
# import CO2 dataset
co2_df = pd.read_csv('./Resources/owid-co2-data.csv')
```

```python
# remove country names that don't represent individual countries

not_country = ['Africa', 'Asia', 'Asia (excl. China & India)','EU-27','EU-28', 'Europe','Europe (excl. EU-27)',
               'Europe (excl. EU-28)', 'International transport', 'Kuwaiti Oil Fires', 'North America',
               'North America (excl. USA)', 'World', 'South America']

co2_df = co2_df.loc[co2_df['country'].apply(lambda x: x not in not_country)]
```

```python
# sort alphabetically
co2_df = co2_df.sort_values(['country', 'year'])
```

```python
#export as clean CSV
co2_df.to_csv('clean_co2_data.csv', index = False)
```

```python
#find which country names are different between dataframes so I can amend the csv names for more country matches in
#future analysis 

def Diff(li1, li2):
    li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2]
    return li_dif

Diff(co2_names, temp_names)
```

```python
# import cleaned resource files

clean_co2 = pd.read_csv('./Resources/global_co2_data_to_2019.csv')
clean_temps = pd.read_csv('./Resources/global_temps_to_2016.csv')
```

```python
# find which columns in co2_df are the least complete, aka have the most NAs

percent_missing = clean_co2.isnull().sum()* 100/len(clean_co2)
percent_missing.sort_values()
```

```python
# drop all columns with over 40% missing values

co2_df = clean_co2.loc[:, clean_co2.isnull().sum()* 100/len(clean_co2) < 40]
co2_df    
```

```python
# keep only most relevant columns

clean_co2 = co2_df.drop(['iso_code','share_global_cumulative_oil_co2', 'cumulative_oil_co2', 'share_global_oil_co2','oil_co2', 'oil_co2_per_capita','share_global_cumulative_coal_co2','cumulative_coal_co2','coal_co2','share_global_coal_co2','coal_co2_per_capita'], axis=1)
clean_co2
```

```python
# drop null values from co2
clean_co2 = clean_co2.dropna()
clean_co2
```

```python
# merge temp and co2 dataframes

co2_temp = pd.merge(clean_co2, clean_temps, how= 'left', on = ['country', 'year'])
co2_temp
```

```python
# reorder columns for my sanity
co2_temp = co2_temp[['iso_code', 'country', 'year', 'temperature(C)', 'population', 'gdp', 'co2', 'share_global_co2', 'co2_per_capita', 'co2_per_gdp', 'co2_growth_prct','co2_growth_abs', 'cumulative_co2', 'share_global_cumulative_co2' ]]
co2_temp
```

```python
# clean nas from combined dataset

co2_temp = co2_temp.dropna()
co2_temp
```

```python
#export as combined CSV
co2_temp.to_csv('global_emissions.csv', index = False)
```
